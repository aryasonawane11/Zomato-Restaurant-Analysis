```python
# 1.How many restaurants offer online order?
df['online_order'].value_counts()
```

```python
 # 2.What is the average rating?
df['rate'].head(10)
```

```python
# Step 2 - remove '/5' and convert to number
df['rate'] = df['rate'].str.split('/').str[0]
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
# Step 3 - now mean will work!
df['rate'].mean()
```

```python
# 3. Which type of restaurant is most common?
df['listed_in(type)'].value_counts()
```

```python
# 4.What is the average cost for two people?
df['approx_cost(for two people)'].mean()
```

```python
# 5.Which are the top 10 highest rated restaurants?
df[['name','rate']].sort_values('rate', ascending=False).head(10)
```

```python
# 6.Which restaurant has the most votes?
df[['name','votes']].sort_values('votes', ascending=False).head(1)
```

```python
# 7. How many restaurants accept online orders vs don't?
df['online_order'].value_counts()
```

```python
# 8.What is the most expensive and cheapest cost for two people?
print('Most Expensive', df['approx_cost(for two people)'].max())
print('Cheapest',df['approx_cost(for two people)'].min())
print('Average Cost',df['approx_cost(for two people)'].mean())
```

```python
# 9.Average rating for restaurants WITH online order vs WITHOUT?
df.groupby('online_order')['rate'].mean()
```

```python
# 10. Which type of restaurant has highest average rating?
df.groupby('listed_in(type)')['rate'].mean().sort_values(ascending=False)
```

```python
# 11. What is the total votes per restaurant type?
df.groupby('listed_in(type)')['votes'].mean().sort_values(ascending=False)
```

```python
# 12. Which restaurant type has the most expensive average cost?
df.groupby('listed_in(type)')['approx_cost(for two people)'].mean().sort_values(ascending=False)
```

```python
# 13. How many restaurants have rating above 4?
df[df['rate']>4].shape[0]
```

```python
# 14. Show only restaurants with rating above 4 AND online order available
df.loc[(df['rate'] > 4) & (df['online_order'] =='Yes')]
```

```python
# 15. What percentage of restaurants offer online ordering?
online = df['online_order'].value_counts()
percentage = (online['Yes'] / len(df)) * 100
print(f'{percentage:.2f}% restaurants offer online ordering')
```

```python
# 16.  Top 5 most voted restaurants with their rating
df[['name','votes','rate']].sort_values('votes', ascending=False).head(5)
```

```python
# 17. Add a new column 'value_for_money'
# If cost < 300 AND rate > 4 = 'High'
# else = 'Low'
df['value_for_money'] = np.where(
    (df['approx_cost(for two people)'] < 300) & (df['rate'] > 4),
    'High',
    'Low'
)
df[['name', 'rate', 'approx_cost(for two people)', 'value_for_money']].head(10)
```

```python
# 18.How many restaurants are High vs Low value for money?
df['value_for_money'].value_counts()
```

```python
# 19. Average rating for High vs Low value restaurants
df.groupby('value_for_money')['rate'].mean()
```

```python
# 20. Save your cleaned and analysed data to CSV
df.to_csv('Zomato_cleaned.csv', index=False)
print('File saved successfully!')
```
